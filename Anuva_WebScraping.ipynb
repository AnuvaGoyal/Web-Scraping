WEBSITE:https://www.shiksha.com/humanities-social-sciences/articles/top-10-colleges-in-india-mhrd-blogId-15781

```python
#IMPORTING THE LIBRARIES

from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.shiksha.com/humanities-social-sciences/articles/top-10-colleges-in-india-mhrd-blogId-15781"
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
soup.title
```

```python
#HEADING

head = soup.find('h1')
print(head.getText())
```

```python
#TO FIND THE NAME OF TOP 10 COLLEGES

h3_list = []
head3 = soup.find_all('h3')
for h in head3:
    h3_list.append(h.getText())
print(h3_list)
```

```python
#TO FIND THE CONTENT STORED IN P TAGS

p_list = []
for text in soup.find_all('p'):
   p_list.append(text.getText())
```

```python
import os      #FOR linesep

#STORING THE CONTENT IN A STRING
complete_text = head.getText()
complete_text+= os.linesep+os.linesep+ h3_list[0]+": " +os.linesep+ p_list[42] + p_list[43]
complete_text+= (os.linesep+os.linesep+ h3_list[1]+": "+os.linesep + p_list[44])
complete_text+= (os.linesep+os.linesep+ h3_list[2]+": "+os.linesep + p_list[45] + p_list[46])
complete_text+= (os.linesep+os.linesep+ h3_list[3]+": "+os.linesep + p_list[47] + p_list[48])
complete_text+= (os.linesep+os.linesep+ h3_list[4]+": "+os.linesep + p_list[49] + p_list[50])
complete_text+= (os.linesep+os.linesep+ h3_list[5]+": "+os.linesep + p_list[51] + p_list[52])
complete_text+= (os.linesep+os.linesep+ h3_list[6]+": "+os.linesep + p_list[53] + p_list[54])
complete_text+= (os.linesep+os.linesep+ h3_list[7]+": "+os.linesep + p_list[55])
complete_text+= (os.linesep+os.linesep+ h3_list[8]+": "+os.linesep + p_list[56] + p_list[57])
complete_text+= (os.linesep+os.linesep+ h3_list[9]+": "+os.linesep + p_list[58])

print(complete_text)
```

```python
#STORING THE CONTENT IN A .CSV FILE

import csv
with open('Anuva_top10collegesinfo.csv', 'w') as f: 
    write = csv.writer(f) 
    write.writerow(complete_text)
```
